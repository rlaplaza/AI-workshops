5. Why do you think you are obtaining this score?

Answer:

<details>
<summary>See answer</summary>

We are trying to build a model using only substrate information, but the ddG values also depend on the catalyst. The problem is that for the same substrate (for example, 1a) we have different ddG values depending on the catalyst (1a–3a = 1.32 and 1a–3b = 1.22). However, the model only sees the substrate, so it receives the same input with two different outputs. This confuses the model, because it lacks the catalyst information needed to distinguish between both cases.

</details>

10. Why do you think you are obtaining this score? Is it better than the models obtained using only the substrate SMILES? Why?

Answer:

<details>
<summary>See answer</summary>

By using two SMILES columns, the model now incorporates information from both the substrate and the catalyst, rather than only the substrate. This resolves the previous issue where the same input (substrate) was associated with different ddG values depending on the catalyst. With both components included, each entry becomes uniquely defined, allowing the model to correctly distinguish between cases such as 1a–3a and 1a–3b and establish a consistent relationship between inputs and outputs.

</details>
